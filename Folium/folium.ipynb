```python
import pandas as pd
import folium
from folium import plugins
```

```python
# Датасет с ресторанами конкретной сети
partizan = pd.read_excel('E://PROJECTS/Python/Folium/Partizan.xlsx')
partizan.head(5)
```

```python
lats = list(partizan.lat)
lons = list(partizan.lon)
places = [[x[0], x[1]] for x in zip(lats, lons)]
```

```python
# Расположение ресторанов на карте
map = folium.Map(places[0], tiles='OpenStreetMap', zoom_start=13)
#plugins.MarkerCluster(places).add_to(map)
plugins.FastMarkerCluster(places).add_to(map)
map
```
